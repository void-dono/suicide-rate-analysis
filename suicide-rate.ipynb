```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('S:\github\Suicide_Rates\master.csv')
df.head()
```

```python
countries=df['country'].unique()
countries
```

```python
# Finding Out Which country has most suicide rate
alpha = 0.7
plt.figure(figsize=(10,25))
sns.countplot(y='country',data=df, alpha=alpha,order = df['country'].value_counts().index)
plt.title('Data by country')
plt.show()
```

```python
# Comparing Suicide rate Male Vs Female

plt.figure(figsize=(16,10))
gender= sns.countplot(x='sex',data=df)
```

```python
# Finding factors of suicide based on the data given
plt.figure(figsize=(16,7))
core = sns.heatmap(df.corr(),annot = True)
```

##### Sizing heat map we can have idea which factors are responsible for suicide. Here we see one posibble factors which is:Population has a impact on suicide rate. Inaddition to that we also see on insight that HDI and GDP has a high corelation.


### Now lets see which Male age group is having high suicide rate

```python
plt.figure(figsize=(20,10))
bar_age = sns.barplot(x='age', y='suicides_no', hue= 'age', data=df, )
```

##### This above plot shows that age group 35-54 years have high chance of commiting suicide. This also provides evidance that people are going through mid life crisis which leds to suicide.

### Now lets see suicide rates based on gender and age_group

```python
plt.figure(figsize=(20,10))
bar_gen_age= sns.barplot(x='sex',y='suicides_no',hue= 'age', data=df)
```
